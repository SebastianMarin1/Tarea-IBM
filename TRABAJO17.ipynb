# Pregunta 1 - Extracción de datos de acciones de Tesla utilizando yfinance (SEBASTIAN MARIN)

```python
import yfinance as yf

# Crear un objeto Ticker para Tesla
tesla = yf.Ticker("TSLA")

# Obtener datos históricos
tesla_stock_data = tesla.history(period="max")

# Mostrar las primeras filas
print(tesla_stock_data.head())
```

# Pregunta 2 - Extracción de datos de ingresos de Tesla utilizando Webscraping (SEBASTIAN MARIN)

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de ingresos de Tesla
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
headers = {"User-Agent": "Mozilla/5.0"}  # Agregar encabezados para evitar bloqueos
html_data = requests.get(url, headers=headers).text

# Parsear los datos HTML
soup = BeautifulSoup(html_data, "html.parser")

# Buscar tablas en la página
tables = soup.find_all("table")

if tables:  # Verificar si se encontraron tablas
    tesla_income_data = pd.read_html(str(tables[0]))[0]  # Extraer la primera tabla
    print("Tabla de ingresos de Tesla extraída con éxito:")
    print(tesla_income_data.head())
else:
    print("No se encontraron tablas en la página web.")
```

# Pregunta 5 - Tablero de acciones e ingresos de Tesla (SEBASTIAN MARIN)

```python
import matplotlib.pyplot as plt

# Gráfico de acciones de Tesla
plt.figure(figsize=(10,5))
plt.plot(tesla_stock_data.index, tesla_stock_data["Close"], label="Precio de Cierre")
plt.title("Historial de Precios de Tesla")
plt.xlabel("Fecha")
plt.ylabel("Precio de Cierre")
plt.legend()
plt.show()
```
